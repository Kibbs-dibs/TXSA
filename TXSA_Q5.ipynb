```python
%pip install transformers 
# Uncomment the line above and run if transformers is not installed

from transformers import BertTokenizer
import string
import nltk
from nltk.corpus import stopwords

# Ensure NLTK resources are available for the filtering comparison
nltk.download('stopwords', quiet=True)

# Global Variable
FILE_PATH = 'Part_A_Dataset\Data_1.txt'
```

```python
# Load the dataset using the global variable
try:
    with open(FILE_PATH, 'r') as f:
        text = f.read()
        print("\n--- Original Text in Data_1 ---")
        print(text)
        print("-" * 30 + "\n")
except FileNotFoundError:
    print(f"Error: {FILE_PATH} not found. Please ensure the file is in the correct directory.")
```

```python
print("--- WordPiece Tokenization ---")

# Initialize the WordPiece tokenizer using a pre-trained base model
# 'bert-base-uncased' automatically handles lowercasing during tokenization
wordpiece_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Execute tokenization using the WordPiece algorithm
wordpiece_tokens = wordpiece_tokenizer.tokenize(text)

print("\nTokenisation using WordPiece Tokenizer ~~~~~~~~~~~~~~~")
print(wordpiece_tokens)
```

```python
print("\n--- Stop Words and Punctuation Filtering (WordPiece) ---")

# Define noise: standard English stopwords + string punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
noise_list = stop_words.union(punctuation)

# Identify stop words present in the specific text
# Note: Since BERT 'uncased' already lowercased the text, we don't need text.lower()
identified_stop_words = sorted(list(set([w for w in wordpiece_tokens if w in stop_words])))

# Filtered Output
filtered_wordpiece_tokens = [w for w in wordpiece_tokens if w not in noise_list]

print(f"\nList of Identified Stop Words in {FILE_PATH}:")
print(identified_stop_words)

print("\nFiltered Output (Noise Removed) ~~~~~~~~~~~~~~~")
print(filtered_wordpiece_tokens)

print(f"\nSummary: Original ({len(wordpiece_tokens)} tokens) -> Filtered ({len(filtered_wordpiece_tokens)} tokens)")
```
