```python
import nltk
from nltk.tokenize import word_tokenize
from textblob import TextBlob

# 1. Input Data
sentence = "The big black dog barked at the white cat and chased away."
tokens = word_tokenize(sentence)

# 2. NLTK POS Tagger
nltk_tags = nltk.pos_tag(tokens)

# 3. TextBlob POS Tagger
blob = TextBlob(sentence)
textblob_tags = blob.tags

# 4. Regular Expression Tagger (Patterns from Lab 05)
patterns = [
     (r'.*ing$', 'VBG'),               # gerunds
     (r'.*ed$', 'VBD'),                # simple past
     (r'.*es$', 'VBZ'),                # 3rd singular present
     (r'.*ould$', 'MD'),               # modals
     (r'.*\'s$', 'NN$'),               # possessive nouns
     (r'.*s$', 'NNS'),                 # plural nouns
     (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),  # cardinal numbers
     (r'.*ment$', 'NN'),               # i.e. wonderment
     (r'.*ful$', 'JJ'),                # i.e. wonderful
     (r'.*', 'NN')                     # nouns (default)
]
regexp_tagger = nltk.RegexpTagger(patterns)
regex_tags = regexp_tagger.tag(tokens)

# Output Results
print("NLTK Tags:", nltk_tags)
print("\nTextBlob Tags:", textblob_tags)
print("\nRegEx Tags:", regex_tags)

# 5. Syntactic Analysis (Parse Tree)
# Defining a simple grammar to handle the sentence structure
grammar = nltk.CFG.fromstring("""
  S -> NP VP
  NP -> DT JJ JJ NN | DT JJ NN | DT NN
  VP -> VBD PP | VBD ADVP | VP CC VP
  PP -> IN NP
  ADVP -> RB
  DT -> 'The' | 'the' | 'a'
  JJ -> 'big' | 'black' | 'white'
  NN -> 'dog' | 'cat'
  VBD -> 'barked' | 'chased'
  IN -> 'at'
  CC -> 'and'
  RB -> 'away'
""")

parser = nltk.ChartParser(grammar)
print("\nParse Tree Output:")
clean_tokens = [t for t in tokens if t.isalnum()]
for tree in parser.parse(clean_tokens):
    tree.pretty_print()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

# Input Text
text = """Classification is the task of choosing the correct class label for a given input. 
In basic classification tasks, each input is considered in isolation from all other inputs, 
and the set of labels is defined in advance. The basic classification task has a number 
of interesting variants. For example, in multiclass classification, each instance may 
be assigned multiple labels; in open-class classification, the set of labels is not 
defined in advance; and in sequence classification, a list of inputs are jointly classified."""

# Initialize the Tokenizer
tokenizer = Tokenizer()

# Fit the tokenizer on the text
tokenizer.fit_on_texts([text])

# Extract tokens (words)
keras_tokens = list(tokenizer.word_index.keys())

print("Keras Tokens List:")
print(keras_tokens)
```
